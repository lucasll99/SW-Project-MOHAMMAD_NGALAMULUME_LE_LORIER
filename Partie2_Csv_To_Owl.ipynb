```python
import pandas as pd
```

### Chargement des csv

```python
df_movies = pd.read_csv('movies.csv', encoding='latin-1',sep = ';')
```

```python
df_actors = pd.read_csv('Actors.csv', encoding='latin-1', sep = ';')
```

```python
df_writers = pd.read_csv('Writers.csv', encoding='latin-1', sep = ';')
```

```python
df_directors = pd.read_csv('Directors.csv', encoding='latin-1', sep = ';')
```

```python
df_theaters= pd.read_csv('Theater.csv', encoding='latin-1', sep = ';')
```

```python
df_movies['genres2'] = df_movies['genres'].apply(lambda x: x.split(','))
```

```python
df_schedules= pd.read_excel('Schedule.xlsx')
```

### Transformation des dataframes en listes de listes:

```python
# transformation dataframe en liste de listes: 
list_movies = df_movies.values.tolist() # movie
list_actors = df_actors.values.tolist() # actors
list_writers = df_writers.values.tolist() # writers
list_directors = df_directors.values.tolist() # directors
list_theaters = df_theaters.values.tolist() # theaters
list_schedules = df_schedules.values.tolist() # schedules 
```

### Avec cette méthode, on obtient au final 4 fichiers contenant respectivement:
1.Déclaration 
2.Les Class Assertion 
3. Les data properties 
4.Les objects properties 

### Conversion Movies en Owl 

```python
f1=open(r'Declaration.txt','w')
f2=open(r'Class_Assertion.txt','w')
f3=open(r'Data_property.txt','w')
c=0
for l in list_movies:
    c=c+1
    # Déclaration individu 
    f1.write('<Declaration>\n\t<NamedIndividual IRI="#'+ str(l[0]) +'"/>\n</Declaration> \n\n')
    # Déclaration des classes:
    # transformation valeur theme en liste, si plusieurs thèmes: 
    test = l[2].split(',')
    if(len(test) == 1): # on écrit la seule classe
        f2.write('<ClassAssertion>\n\t<Class IRI="#'+ test[0] +'"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n</ClassAssertion>\n\n')

    if(len(test) == 2): # on écrit les deux classes
        f2.write('<ClassAssertion>\n\t<Class IRI="#'+ test[0] +'"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n</ClassAssertion>\n')
        f2.write('<ClassAssertion>\n\t<Class IRI="#'+ test[1] +'"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n</ClassAssertion>\n\n')

    if(len(test) == 3): # on écrit les trois classes
        f2.write('<ClassAssertion>\n\t<Class IRI="#'+ test[0] +'"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n</ClassAssertion>\n')
        f2.write('<ClassAssertion>\n\t<Class IRI="#'+ test[1] +'"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n</ClassAssertion>\n')
        f2.write('<ClassAssertion>\n\t<Class IRI="#'+ test[2] +'"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n</ClassAssertion>\n\n')

    # Déclaration des attributs de l'individu
    # titre l[0], marche pas encore, je me demande si ce n'est pas à cause du kill bill (volume 1) erreur utf-8
    f3.write('<DataPropertyAssertion>\n\t<DataProperty IRI="#title"/>\n\t<NamedIndividual IRI="#'+ str(l[0]) +'"/>\n\t<Literal>' + str(l[1]) + '</Literal>\n</DataPropertyAssertion>\n\n')
    
    # country l[4]
    f3.write('<DataPropertyAssertion>\n\t<DataProperty IRI="#country"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n\t<Literal>' + str(l[4]) + '</Literal>\n</DataPropertyAssertion>\n\n')
    
    # language l[5]
    f3.write('<DataPropertyAssertion>\n\t<DataProperty IRI="#language"/>\n\t<NamedIndividual IRI="#'+ l[0] +'"/>\n\t<Literal>' + str(l[5]) + '</Literal>\n</DataPropertyAssertion>\n\n')
    
    #durée l[6]: 
    f3.write('<DataPropertyAssertion>\n\t<DataProperty IRI="#duration"/>\n\t<NamedIndividual IRI="#'+ str(l[0]) +'"/>\n\t<Literal datatypeIRI="http://www.w3.org/2001/XMLSchema#positiveInteger">' + str(l[6]) + '</Literal>\n</DataPropertyAssertion>\n\n')    
    
    # Déclaration object property # il n'y en a pas, se fait à partir des dataframes actors, writers, directors et theaters 
    #f2.write(l)
f1.close()
f2.close()
f3.close()
```
